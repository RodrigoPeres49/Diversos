# - Convers√£o do arquivo csv em bando de dados.

```python
import pandas as pd
import sqlite3

arquivo = 'base_varejo.csv'
df = pd.read_csv(arquivo, sep=';')


conn = sqlite3.connect('varejo.db')


df.to_sql('produtos_varejo', conn, if_exists='replace', index=False)

conn.close()
```

# Algumas consultas em sql

# 1- Filtrar produtos por categoria

```python
import sqlite3
from tabulate import tabulate

conn = sqlite3.connect('varejo.db')

cursor = conn.cursor()

def contagem_categoria(cursor):

    cursor.execute(" SELECT PR_CAT categoria, COUNT(*) contagem FROM produtos_varejo GROUP BY PR_CAT  ORDER BY DATA")
    
    linhas = cursor.fetchall()
    
    print("Lista de Produtos")
    
    colunas = [description[0] for description in cursor.description]
    print (tabulate(linhas, headers= colunas, tablefmt='grid'))

contagem_categoria(cursor)
    
conn.close
```
